```python
import random
numero = random.randint(1,50)
tentativas = 5
i = 0
print('Esse é o jogo da adivinhação!')
print('Você tem {tentativas} 5 tentativas para adivinhar o número entre 1 a 50')
while i < 5:
  palpite = int(input('Qual é o seu palpite?'))
  if palpite == numero:
    print('Parabéns você acertou o numero secreto!')
    break
  else:
    tentativas -= 1
  if palpite < numero:
    print('O número secreto é maior. Tente novamente!')
  else:
    print('O número secreto é menor. Tente novamente!')
  if tentativas > 0:
    print(f'Você ainda tem {tentativas} tentativas.')
  else:
    print(f'Você não conseguiu adivinhar o número. O número secreto era {numero}.')
    break
```
